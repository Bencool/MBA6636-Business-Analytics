## Set 1 Automobile Data Set (Bin Liao)


````python
import pandas as pd
import numpy as np
automobile = pd.read_csv(
    "https://raw.githubusercontent.com/dearbharat/pandas/main/automobile.csv")
automobile.head()
````

1. Print most expensive car’s company name and price.


````python
# Using the idxmax() function to findout the index that contain the maximun value. Then use the index to find out the value of the maximum value.
max_price_index = automobile['price'].idxmax()
most_expensive_car = automobile.loc[max_price_index, ['company', 'price']]
print("Most expensive car:")
print(most_expensive_car)
````

2. Display the count of the number of cars for each company.


````python
# Using the value_counts() function in pandas' to count the number of occurrences of each unique value in a column.
car_count = automobile['company'].value_counts()
print("\nNumber of occurrences of each unique value in the 'company' column:\n")
car_count.head(20)
````

3. Print the least expensive car in each of the company.


````python
# Using the group_by function on "company" and then use the idxmin() function to find the index of the minimun value on price.
least_expensive_cars = automobile.loc[automobile.groupby('company')['price'].idxmin()]

print(least_expensive_cars.to_string(index=False))

print("\nThe least expensive cars by company are:\n")
for company, car in least_expensive_cars.iterrows():
    print("Company:", car['company'])
    print("Car Index:", car['index'])
    print("Price:", car['price'])
    print()
````

4. Display the number of cars by num-of-cylinders for each body-style.
   Hint: Use Group_by or Pivot function


````python
# Using the groupby() function in combination with the count() function.
car_counts = automobile.groupby(['body-style', 'num-of-cylinders']).size().reset_index(name='count')
print("Display the number of cars by num-of-cylinders for each body-style.\n")
car_counts.head(20)
````

## Set 2 Bank Marketing (Ziyin Xu)


````python
import pandas as pd
bank = pd.read_csv(
    "https://raw.githubusercontent.com/dearbharat/pandas/main/bank_marketing.csv", nrows=1000)
bank.head(1)
````

1. Display all the unique values in the column "job".


````python
unique_jobs = bank['job'].unique()
print("Unique jobs:\n")
print(unique_jobs)
````

2. Create a cross table with education as index column, job as columns and mean of age as values.


````python
import numpy as np
cross_table = pd.pivot_table(data=bank, index='education', columns='job', values='age', aggfunc=np.mean)
print(cross_table)
````

3. Create a Pivot table of the average of the duration with job as column, marital and housing as index and average of duration as values.


````python
import numpy as np
pivot_table = pd.pivot_table(bank, values='duration', index=['marital', 'housing'], columns='job', aggfunc=np.average)
print(pivot_table)
````

2. Display the average Runtime for each Genre. Do not try to normalize the Genre column. Use it as is.

3. Display the top three movies by metascore in each of the year. Display the Title, Director, and Revenue in the output dataframe.


````python
# Define a function to get the top three movies by Metascore for each year
def get_top_three_movies(group):
    return group.sort_values(by='Metascore', ascending=False).head(3)

# Apply the function to the IMDB dataset
top_movies_by_year = IMDB.groupby('Year').apply(get_top_three_movies)[['Year', 'Title', 'Director', 'Revenue (Millions)']]

top_movies_by_year = top_movies_by_year.reset_index(drop=True)

# Print the resulting dataframe
top_movies_by_year
````
